```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
import pandas as pd
import matplotlib.ticker as mticker

plt.rcParams['font.sans-serif'] = ['SimHei']  # 适用于 Windows/Linux
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
```

```python
import pandas as pd

# Read Excel
file_path = "C:/Users/dell/PythonDataAnalysis/wdn/即时对抗性反杀.xlsx"
xls = pd.ExcelFile("C:/Users/dell/PythonDataAnalysis/wdn/即时对抗性反杀.xlsx")

xls.sheet_names
```

```python
# 读取第一个工作表的数据
df = pd.read_excel(xls, sheet_name='Sheet1')

# 显示数据的前几行，以了解其结构
df.head()
```

```python
# 按案件数排序，方便可视化
df = df[df['独特省份'].apply(lambda x: isinstance(x, str))]
df_sorted = df.sort_values(by='案件数', ascending=False)

# 设置柱状图的宽度
bar_width = 0.4

# 创建位置索引
x_indexes = np.arange(len(df_sorted['独特省份']))

# 创建画布和双轴
fig, ax1 = plt.subplots(figsize=(12, 6))

# 绘制柱状图（并排显示）
ax1.bar(x_indexes - bar_width/2, df_sorted['案件数'], color='grey', width=bar_width, label='案件数', alpha=0.7)
ax1.bar(x_indexes + bar_width/2, df_sorted['重型案件'], color='orange', width=bar_width, label='重型案件', alpha=0.7)
ax1.set_xlabel("省份")
ax1.set_ylabel("案件数量", color='black')
ax1.tick_params(axis='y', labelcolor='black')
ax1.set_xticks(x_indexes)
ax1.set_xticklabels(df_sorted['独特省份'], ha='right')
ax1.yaxis.set_major_locator(mticker.MaxNLocator(integer=True))

# 第二个y轴（重刑率折线图）
ax2 = ax1.twinx()
ax2.plot(x_indexes, df_sorted['比率'], color='black', marker='o', linestyle='-', label='重刑率')
ax2.set_ylabel("重刑率", color='black')
ax2.tick_params(axis='y', labelcolor='black')
ax2.set_ylim(0,ax2.get_ylim()[1])

# 标题和图例
plt.title("各省即时对抗性反杀案件数量及重刑率")
ax1.legend(loc='upper left', bbox_to_anchor=(0.15, 1), fontsize=10, frameon=True)
ax2.legend(loc='upper left', bbox_to_anchor=(0.15, 0.9), fontsize=10, frameon=True)

# 显示图表
plt.savefig("即时对抗性反杀.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
###
import pandas as pd
import statsmodels.api as sm

# 处理“防卫因素”为哑变量
dftest = pd.get_dummies(dftest, columns=['防卫因素'], prefix='防卫', drop_first=True)

# 处理“被害人过错”为哑变量
dftest['被害人过错_重大'] = (dftest['被害人过错'] == 1).astype(float)
dftest['被害人过错_具有'] = (dftest['被害人过错'] == 2).astype(float)
dftest['被害人过错_明显'] = (dftest['被害人过错'] == 3).astype(float)
dftest.drop(columns=['被害人过错'], inplace=True)

#for col in df.columns:
#    if col != '案号':  # 跳过字符串列
#        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# **OLS 回归：被害人过错对量刑刑期的影响**
X = dftest[['被害人过错_重大', '被害人过错_具有', '被害人过错_明显', '防卫_2', '防卫_3', '防卫_4']]
X = sm.add_constant(X)
Y = dftest['量刑刑期']

# 运行 OLS
ols_model = sm.OLS(Y, X).fit()
print(ols_model.summary())
```
